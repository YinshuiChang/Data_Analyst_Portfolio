# Load and visualize the data

```python

# data_precip_y = np.load("../data/WaterPrecip_datacube_CNN_y.npy")
data_precip_y = np.load("../data/WaterPrecip_datacube_CNN_y_s2.npy")
```

```python
temp_1 = np.load("./Testing/cnn_pred_IIOII_v01_v57.npy")
temp_2= np.load("./Testing/cnn_pred_IIIO_v01_v57.npy")
```

```python
temp = temp_1
nshow = 5
fig, axs = plt.subplots(nrows=nshow, ncols=2, sharex=True, sharey=True, figsize=(10,10))
fig.tight_layout()
for i in range(nshow):
    sns.heatmap(data_precip_y[i,::-1,:,0], cmap="viridis", ax=axs[i,0])
    sns.heatmap(temp[i,::-1,:,0], cmap="viridis", ax=axs[i,1])
axs[0,0].set_title('original data')
axs[0,1].set_title('predicted data')
plt.show()
```

```python
nshow = 5
fig, axs = plt.subplots(nrows=nshow, ncols=2, sharex=True, sharey=True, figsize=(10,10))
fig.tight_layout()
for i in range(nshow):
    sns.heatmap(temp_1[i,::-1,:,0], cmap="viridis", ax=axs[i,0])
    sns.heatmap(temp_2[i,::-1,:,0], cmap="viridis", ax=axs[i,1])
axs[0,0].set_title('predicted data_1')
axs[0,1].set_title('predicted data_2')
plt.show()
```
