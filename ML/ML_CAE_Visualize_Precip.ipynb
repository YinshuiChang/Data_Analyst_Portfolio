# Load and visualize the data

```python
datacube_precip = np.load("../data/WaterPrecip_datacube_CAE_testing.npy")
```

```python
temp_1 = np.load("./Testing/temp_vx01_v34.npy")
temp_2 = np.load("./Testing/temp_CVAE_v01.npy")
```

```python
temp = temp_1
idx = 0
fig, axs = plt.subplots(nrows=4, ncols=2, sharex=True, sharey=True, figsize=(10,8))
fig.tight_layout()
for i in range(4):
    sns.heatmap(datacube_precip[i,::-1,:,idx], cmap="viridis", ax=axs[i,0])
    sns.heatmap(temp[i,::-1,:,idx], cmap="viridis", ax=axs[i,1])
axs[0,0].set_title('original Data')
axs[0,1].set_title('reconstructed Data')
plt.show()
```

```python
idx = 0
fig, axs = plt.subplots(nrows=4, ncols=2, sharex=True, sharey=True, figsize=(10,8))
fig.tight_layout()
for i in range(4):
    sns.heatmap(temp_1[i,::-1,:,idx], cmap="viridis", ax=axs[i,0])
    sns.heatmap(temp_2[i,::-1,:,idx], cmap="viridis", ax=axs[i,1])
axs[0,0].set_title('reconstructed Data_1')
axs[0,1].set_title('reconstructed Data_2')
```
